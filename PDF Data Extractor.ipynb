```python
import pdfplumber
from docx import Document

# Extract text from PDF with pdfplumber
pdf_file_path = "c4611.pdf"
try:
    with pdfplumber.open(pdf_file_path) as pdf:
        pdf_text = ""
        for page in pdf.pages:
            pdf_text += page.extract_text() + "\n"
except Exception as e:
    print(f"Error extracting text from PDF: {e}")
    pdf_text = ""

# Create a Word document and save the extracted text
if pdf_text:
    document = Document()
    for line in pdf_text.splitlines():
        document.add_paragraph(line)
    document.save("my.docx")
    print("Text successfully saved to my.docx")
else:
    print("No text extracted to save.")
```
